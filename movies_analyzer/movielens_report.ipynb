## &emsp;&emsp;&emsp;&emsp; Audience Ratings & Engagement

### Next, using Ratings.Movies, examine how ratings are distributed by year (dist_by_year)

```python
%timeit Graph().text_histogram(Ratings().Movies().dist_by_year(),100,0)
rt_by_year = Ratings().Movies().dist_by_year()
histogram = Graph().text_histogram(rt_by_year,100,0)
histogram
```

### The histogram shows three major peaks of users' activity amongst the years analyzed: 2000, 2007 and 2017. The first peak could be explained by the popularity of the MovieLens platform in the early 2000s. The second peak could be associated with growing interest in online cinemas and referral services in the mid-2000s, whereas the third one is probably linked with increase in development of streaming services (like Netflix).

### Despite all these explainations, we could conclude that user engagement on MovieLens was significantly influenced by multiple interdependent factors: technological infrastructure evolution, cultural and cinematic Trends, etc.

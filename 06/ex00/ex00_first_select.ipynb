## get the subtable using only one query where:

### only uid and datetime are used

```python
only_uid_and_datetime = pd.io.sql.read_sql_query("SELECT uid, datetime FROM pageviews", connection)
only_uid_and_datetime
```

### only user data (user_*) is used and not admin data

```python
#var1
pageviews_only_user_data = pd.io.sql.read_sql_query("SELECT uid, datetime FROM pageviews WHERE uid NOT LIKE 'admin_%'", connection)
pageviews_only_user_data
```

```python
#var2
pageviews_only_user_data = pd.io.sql.read_sql_query("SELECT uid, datetime FROM pageviews WHERE uid LIKE 'user_%'", connection)
pageviews_only_user_data
```

### it is sorted by uid in ascending order

```python
sorted_by_uid = pd.io.sql.read_sql_query("SELECT uid, datetime FROM pageviews WHERE uid LIKE 'user_%' ORDER BY uid ASC", connection)
sorted_by_uid
```

### datetime is converted to DatetimeIndex

```python
datetime = pd.io.sql.read_sql_query("SELECT uid, datetime FROM pageviews WHERE uid NOT LIKE 'admin_%' ORDER BY uid ASC", connection, parse_dates=['datetime'], index_col='datetime')
datetime
```
